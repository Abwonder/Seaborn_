### Seaborn 

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris') ## loading dataset from within seaborn
```

```python
iris
```

```python
## Normal scatter plot with no attributes attached
sns.scatterplot(x = iris.sepal_length , y = iris.sepal_width)
```

```python
### Changed the Y axis!!!
sns.scatterplot(x = iris.sepal_length , y = iris.petal_length)
```

```python
### distribution plot (equivalent to histgram)

sns.displot(iris['sepal_length'])
```

```python
## the same thing as the one above
sns.histplot(iris.sepal_length)  ## column can also the accessed this way
```

```python
## the same thing as the one above
sns.histplot(iris['sepal_length'])
```

```python
tips = sns.load_dataset('tips')#####  loading another data set title == tips
```

```python
tips
```

```python
tips.head() #### print out the first five columns
```

```python
sns.scatterplot(x = tips.total_bill , y = tips.tip)  ##scatterplot normala
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()  ## categorical value count
```

```python
tips['size'].value_counts()  ## match uniques values with their freq
```

```python
## way of plotting relative data set for comparation within the same col
#Note: data and hue parameters added(needed)
sns.relplot(x = tips.total_bill , y = tips.tip , data = tips,hue = 'smoker' )
```

```python
sns.relplot(x = tips.total_bill , y = tips.tip , data = tips, hue = 'size')
```

```python
## Same example as above using style 
sns.relplot(x = tips.total_bill , y = tips.tip , data = tips,style=  'size' )
```

```python
# Combined usage for hue and style
sns.relplot(x = tips.total_bill , y = tips.tip , data = tips,style=  'size',hue = 'time' )
```

```python
tips
```

```python
## categorical plot, similar to bar plot but show cluster of the data rep
sns.catplot(x = 'day' , y = 'total_bill' , data = tips )
```

```python
sns.catplot(x = 'smoker' , y = 'tip' , data = tips )
```

```python
## this combines histplot with scatteplot
sns.jointplot(x = tips.total_bill , y = tips.tip)
```

```python
# data summary for all the numerical dataset
sns.pairplot(tips)
```

### Question 1
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. 

Below are five plots type in seaborn library :

__Distribution plots__

This type of plot is used for examining both types of distributions, i.e., univariate and bivariate distribution.

__Relational plots:__

This type of plot is used to understand the relation between the two given variables.

__Categorical plots__

This type of plot is used to visualize the distribution of categorical variables.

__Regression plots__


This type of plot is used to visualize the relationship between two variables and their regression line.

__Matrix plots__

This type of plot is used to visualize the correlation between multiple variables in a dataset.

### Question 2

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.


```python
import seaborn as sn

fmri = sn.load_dataset("fmri")
fmri
```

```python
sn.lineplot(x = fmri.timepoint, y = fmri.signal, data = fmri, style = "event", hue = "region")
```

### Question 3

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
### Impporting needed library
import seaborn as sns

# Loading dataset
titanic = sns.load_dataset('titanic')
# show data
titanic
```

```python
# plotting
sns.boxplot(x='pclass', y='age', data=titanic)
```

```python
sns.boxplot(x='pclass', y='fare', data=titanic)
```

### Question 4
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
# import library
import seaborn as sns
import matplotlib.pyplot as plt  ## added this to be able to maniputa fig

diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
## ploting histogram 
plt.figure(figsize = (10, 6))  ## manipulate the dimension fo the graph
sns.histplot(x = diamonds["price"], data = diamonds, hue = "cut")
plt.show()
```

### Question 5
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
## libraries
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
```

```python
sns.pairplot(data = iris, hue = "species")
plt.show()
```

### Question 6
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
flights = sns.load_dataset('flights')
flights
```

```python
## building pivot table for the dataset from all the column as shown above
flights = flights.pivot('month', 'year', 'passengers')

## heatmap
sns.heatmap(flights, annot=True, fmt='d')
plt.show()
```
